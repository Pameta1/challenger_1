```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

####Exploración inicial de los datos

```python
tienda2 = pd.read_csv(url2)
tienda2.head()
```

```python
tienda3 = pd.read_csv(url3)
tienda3.head()
```

```python
tienda4 = pd.read_csv(url4)
tienda4.head()
```

```python
tienda.columns
```

```python
tienda.info()
```

```python
tienda_list = [tienda, tienda2, tienda3, tienda4]
tienda_names = ['tienda_1', 'tienda_2', 'tienda_3', 'tienda_4']

all_tiendas = pd.DataFrame()
```

```python
for i, df in enumerate(tienda_list):
    df_copy = df.copy() # Create a copy to avoid SettingWithCopyWarning
    df_copy['tienda'] = tienda_names[i]
    all_tiendas = pd.concat([all_tiendas, df_copy], ignore_index=True)
```

```python
ingresos_por_tienda = all_tiendas.groupby('tienda')['Precio'].sum().reset_index()
ingresos_por_tienda.columns = ['Tienda', 'Ingreso Total']
print(ingresos_por_tienda)
```

```python
ingresos_por_tienda['Ingreso Total'] = ingresos_por_tienda['Ingreso Total'].astype(int)
```

```python
print(ingresos_por_tienda.to_string(index=False))
```

```python
tienda['Precio']
```

```python
suma_de_precios = 0
for precio in tienda['Precio']:
  suma_de_precios += precio
print(suma_de_precios)
```

# 2. Ventas por categoría

```python
ventas_por_categoria = all_tiendas.groupby(['tienda', 'Categoría del Producto']).size().reset_index(name='Cantidad de Ventas')
print(ventas_por_categoria)
```

####Categorías ordenadas más populares de cada tienda

```python
ventas_ordenadas = ventas_por_categoria.sort_values(['tienda', 'Cantidad de Ventas'], ascending=[True, False])
print(ventas_ordenadas)
```

####Categorías más vendidas por tienda

```python
categorias_top = ventas_ordenadas.groupby('tienda').first().reset_index()
print(categorias_top)
```

```python
ventas_por_categoria = all_tiendas.groupby(['tienda', 'Categoría del Producto']).size().reset_index(name='Cantidad de Ventas')
```

```python
pivot_data = ventas_por_categoria.pivot(index='Categoría del Producto', columns='tienda', values='Cantidad de Ventas').fillna(0)
```

```python
plt.figure(figsize=(12,6))

for tienda in pivot_data.columns:
    plt.plot(pivot_data.index, pivot_data[tienda], marker='o', label=tienda)

plt.title('Cantidad de productos vendidos por categoría en cada tienda')
plt.xlabel('Categoría del Producto')
plt.ylabel('Cantidad de Ventas')
plt.xticks(rotation=45)
plt.legend(title='Tienda')
plt.tight_layout()
plt.show()
```

# 3. Calificación promedio de la tienda


```python
calificacion_promedio = all_tiendas.groupby('tienda')['Calificación'].mean().reset_index()
calificacion_promedio.columns = ['Tienda', 'Calificación Promedio']
print(calificacion_promedio)
```

```python
calificacion_promedio['Calificación Promedio'] = calificacion_promedio['Calificación Promedio'].round(2)
print(calificacion_promedio)
```

# 4. Productos más y menos vendidos

```python
ventas_por_producto = all_tiendas.groupby(['tienda', 'Producto']).size().reset_index(name='Cantidad de Ventas')
print(ventas_por_producto)
```

```python
ventas_ordenadas = ventas_por_producto.sort_values(['tienda', 'Cantidad de Ventas'], ascending=[True, False])
print(ventas_ordenadas)
```

```python
top_productos = ventas_ordenadas.groupby('tienda').head(5).reset_index(drop=True)
print(top_productos)
```

####Los productos más vendidos de cada tienda

```python
bottom_productos = ventas_ordenadas.groupby('tienda').tail(5).reset_index(drop=True)
print(bottom_productos)
```

```python
top_productos = ventas_por_producto.sort_values(['tienda', 'Cantidad de Ventas'], ascending=[True, False])
top_5_por_tienda = top_productos.groupby('tienda').head(5).reset_index(drop=True)
```

```python

plt.figure(figsize=(12,6))
sns.scatterplot(data=top_5_por_tienda, x='Producto', y='Cantidad de Ventas', hue='tienda', s=150)

plt.title('Top 5 productos más vendidos por tienda')
plt.xlabel('Producto')
plt.ylabel('Cantidad de Ventas')
plt.xticks(rotation=45)
plt.legend(title='Tienda')
plt.tight_layout()
plt.show()
```

####Productos menos vendidos

```python
bottom_5_por_tienda = ventas_por_producto.sort_values(['tienda', 'Cantidad de Ventas'], ascending=[True, True])
bottom_5_por_tienda = bottom_5_por_tienda.groupby('tienda').head(5).reset_index(drop=True)
print(bottom_5_por_tienda)
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(data=bottom_5_por_tienda, x='Producto', y='Cantidad de Ventas', hue='tienda', s=150)

plt.title('Top 5 productos menos vendidos por tienda')
plt.xlabel('Producto')
plt.ylabel('Cantidad de Ventas')
plt.xticks(rotation=45)
plt.legend(title='Tienda')
plt.tight_layout()
plt.show()
```
